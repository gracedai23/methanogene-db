GFF File Format: 
- Fields must be tab-separated. Also, all but the final field in each feature line must contain a value; "empty" columns should be denoted with a '.'
[0] seqname - name of the chromosome or scaffold; chromosome names can be given with or without the 'chr' prefix. Important note: the seqname must be one used within Ensembl, i.e. a standard chromosome name or an Ensembl identifier such as a scaffold ID, without any additional content such as species or assembly. See the example GFF output below.
[1] source - name of the program that generated this feature, or the data source (database or project name)
[2] feature - feature type name, e.g. Gene, Variation, Similarity
[3] start - Start position* of the feature, with sequence numbering starting at 1.
[4] end - End position* of the feature, with sequence numbering starting at 1.
[5] score - A floating point value.
[6] strand - defined as + (forward) or - (reverse).
[7] frame - One of '0', '1' or '2'. '0' indicates that the first base of the feature is the first base of a codon, '1' that the second base is the first base of a codon, and so on..
[8] attribute - A semicolon-separated list of tag-value pairs, providing additional information about each feature.

```python
#create an empty dictionary to store the gene counts for each species
import pandas as pd
gene_counts = {}

for path_name in faa_paths:
    species = path_name[:15]  #keep species name from the path_name
    gene_counts[species] = {}

    #read the GFF file
    with open(path_name, 'r') as file:
        for line in file:
            if line.startswith('>'):
                if "mrtA" in line or "MrtA" in line:
                    if 'mrtA' not in gene_counts[species]:
                        gene_counts[species]['mrtA'] = 1
                    else:
                        gene_counts[species]['mrtA'] += 1
                if "mrtB" in line or "MrtB" in line:
                    if 'mrtB' not in gene_counts[species]:
                        gene_counts[species]['mrtB'] = 1
                    else:
                        gene_counts[species]['mrtB'] += 1
                if "mrtG" in line or "MrtG" in line:
                    if 'mrtG' not in gene_counts[species]:
                        gene_counts[species]['mrtG'] = 1
                    else:
                        gene_counts[species]['mrtG'] += 1


# Create a dataframe from the gene counts
df = pd.DataFrame.from_dict(gene_counts, orient='index')
df.fillna(0, inplace=True)
```

```python
#the below section is to extract just genes that are mcr + mrt from species that have at least 1 mrt gene 
# later, we will do an analysis of species that only have mcr vs have mrt & mcr. also are there species that have 2 mcr and no mrt? need to see if it is rly mrt or not!
# expect that there are a lot more genes that are actually mcrA but aren't marked so


import pandas as pd
#read the text file of gene names into a list
gene_file = open("jgi_ncbi_downloads.txt", "r")
data = gene_file.read()
target_strings = data.split("\n")
print(target_strings)
gene_file.close()
```

```python
import os

# Initialize a list to store the paths of matching GFF files
gff_paths = ['./GCA_001477655.1/GCF_001477655.1_ASM147765v1_genomic.gff', './GCA_001563245.1/GCF_001563245.1_ASM156324v1_genomic.gff', 
             './GCA_000762265.1/GCF_000762265.1_ASM76226v1_genomic.gff', './GCA_000012545.1/GCF_000012545.1_ASM1254v1_genomic.gff', './GCA_902762095.1/GCA_902762095_genomic.gff']

# Walk through the directory and its subdirectories
for root, _, files in os.walk("."):
    for filename in files:
        for target_string in target_strings:
            # Check if the target string is in the filename
            if target_string in filename:
                # Check if the file is a GFF file (you can adjust this condition as needed)
                if filename.endswith(".gff"):
                    # Construct the full path to the GFF file
                    gff_path = os.path.join(root, filename)
                    gff_paths.append(gff_path)
print(len(gff_paths))
# Print the list of matching GFF file paths
for path in gff_paths:
    print(path)
```

```python
#create dataframe to store the output 
df = pd.DataFrame(columns=["assembly_accession", "index", "gene", "start", "stop"])

# iterate through the GFF files
for path_name in gff_paths:
    species = path_name[2:17]  #keep species name from the path_name
    gene_count = 0            #index to keep count of number of times gene shows up in one species
    
    with open(path_name, "r") as file:
        for line in file:
            if not line.startswith('#'):
                fields = line.strip().split('\t')
                if "Name=mrtG" in line or "Name=MrtG" in line:   #specify desired gene
                    start = int(fields[3])
                    stop = int(fields[4])
                    gene_count += 1
                    new_row = pd.DataFrame({"assembly_accession": [species],
                                            "index": [gene_count],
                                            "gene": ['mrtG'],
                                            "start": [start],
                                            "stop": [stop]})
                    df = pd.concat([df, new_row], ignore_index=True)
                if "Name=mcrG" in line or "Name=McrG" in line:   #specify desired gene
                    start = int(fields[3])
                    stop = int(fields[4])
                    gene_count += 1
                    new_row = pd.DataFrame({"assembly_accession": [species],
                                            "index": [gene_count],
                                            "gene": ['mcrG'],
                                            "start": [start],
                                            "stop": [stop]})


                    df = pd.concat([df, new_row], ignore_index=True)

df.head(n=10)
```

```python
# Function to find fasta file in a folder and extract the gene sequence
def get_sequence(folder_name, start, stop):
    fasta_file = [file for file in os.listdir(folder_name) if file.endswith(".fna")]
    print(fasta_file)
    if not fasta_file:
        return None
    fasta_path = os.path.join(folder_name, fasta_file[0])
    with open(fasta_path, "r") as file:
        lines = file.readlines()
        sequence = "".join(line.strip() for line in lines[1:] if not line.startswith(">"))
    return sequence[start-1:stop]
```

```python
sequences = []

for _, row in df.iterrows():
    folder_name = row["assembly_accession"]
    start, stop = row["start"], row["stop"]
    sequence = get_sequence(folder_name, start, stop) #extract the gene sequence using function
    sequences.append(sequence)

# add column of sequences to existing dataframe
df["Sequence"] = sequences
df
```

```python
#out_fasta should be "name.fasta" to ensure the output file is in fasta format
def to_fasta(df, out_fasta):
    df["organism_name"] = df["organism_name"].str.replace(" ","-")
    df["strain"] = df["strain"].str.replace(" ","-")
    with open(out_fasta, 'w') as file:
        for _, row in df.iterrows():
            header = f'>{row["assembly_accession"]}|{row["organism_name"]}|{row["strain"]}|{row["gene"]}|{row["start"]}|{row["stop"]}'
            sequence = row["Sequence"]
            file.write(f'{header}\n{sequence}\n')
    file.close()
```

```python
#export to local fasta file
to_fasta(df, "mcrG_mrtG.fasta")
```
