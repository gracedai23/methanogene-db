```python
# Function to find fasta file in a folder and extract the gene sequence
def get_sequence(folder_name, start, stop):
    fasta_file = [file for file in os.listdir(folder_name) if file.endswith(".fna")]
    if not fasta_file:
        return None
    fasta_path = os.path.join(folder_name, fasta_file[0])
    with open(fasta_path, "r") as file:
        lines = file.readlines()
        sequence = "".join(line.strip() for line in lines[1:] if not line.startswith(">"))
    return sequence[start-1:stop] 
```

# Output dataframes to a formatted fasta file 

```python
#out_fasta should be "name.fasta" to ensure the output file is in fasta format
def to_fasta(df, out_fasta):
    df["organism_name"] = df["organism_name"].str.replace(" ","-")
    df["strain"] = df["strain"].str.replace(" ","-")
    with open(out_fasta, 'w') as file:
        for _, row in df.iterrows():
            header = f'>{row["assembly_accession"]}|{row["organism_name"]}|{row["strain"]}|{row["gene"]}|{row["start"]}|{row["stop"]}|{row["orientation"]}'
            sequence = row["Sequence"]
            file.write(f'{header}\n{sequence}\n')
    file.close()
```

```python
import os
import re 
import pandas as pd

id_df = pd.DataFrame(columns=['gene_id', 'gene', 'start', 'stop'])

def extract_gene_id(line):
    match = re.search(r'gene_id "([^"]+)"', line)
    if match:
        return match.group(1)
    else:
        return None

def to_gtf(df, folder_name, out_gtf):
    with open(out_gtf, 'w') as out_file:
        index = 0
        for _, row in df.iterrows():
            assembly_accession = row['assembly_accession']
            gtf_file = [file for file in os.listdir(folder_name) if assembly_accession[3:14] in file]
            
            if not gtf_file:
                print(f"No matching GTF file for {assembly_accession}")
                continue
            
            gtf_path = os.path.join(folder_name, gtf_file[0])
            start = row['start']
            stop = row['stop']
            gene = row['gene']
            
            matched = False 
            with open(gtf_path, 'r') as gtf_file:
                for line in gtf_file:
                    if not line.startswith("#"):
                        fields = line.strip().split('\t')
                        gene_id = extract_gene_id(line)
                  
                        if fields[3] == str(start) and fields[4] == str(stop):
                            out_file.write(line)
                            id_df.loc[index, 'gene_id'] = gene_id 
                            id_df.loc[index, 'start'] = start
                            id_df.loc[index, 'stop'] = stop 
                            id_df.loc[index, 'gene'] = gene 
                            matched = True
                            index+=1
                        elif fields[2] == str(start) and fields[3] == str(start):
                            out_file.write(line)
                            id_df.loc[index, 'gene_id'] = gene_id 
                            id_df.loc[index, 'start'] = start
                            id_df.loc[index, 'stop'] = stop 
                            id_df.loc[index, 'gene'] = gene 
                            matched = True
                            index+=1
                        
            if not matched:
                print(f"No matching entry for {assembly_accession} and gene {row['gene']} {start} {stop}")
    file.close()
```

```python
import os

def get_complement(sequence):
    #Returns the complement of a DNA sequence
    complement = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    return "".join(complement[base] for base in sequence)

def get_sequence(folder_name, start, stop, orientation):
    fasta_file = [file for file in os.listdir(folder_name) if file.endswith(".fna")]
    if not fasta_file:
        return None
    fasta_path = os.path.join(folder_name, fasta_file[0])
    
    with open(fasta_path, "r") as file:
        lines = file.readlines()
        sequence = "".join(line.strip() for line in lines[1:] if not line.startswith(">"))

    extracted_sequence = sequence[start-1:stop]

    #if the orientation is negative, reverse and complement the sequence
    if orientation == "-":
        extracted_sequence = get_complement(extracted_sequence[::-1])

    return extracted_sequence
```

```python
def filter_fasta(input_fasta, output_fasta):
    with open(input_fasta, 'r') as infile, open(output_fasta, 'w') as outfile:
        keep = False
        for line in infile:
            if line.startswith(">"):
                keep = gene in line
            if keep:
                outfile.write(line)
```

```python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord 
```

```python
#need to match with gff files and take the reverse of the sequence, then translate that as well...

def to_protein(input_fasta, output_fasta):
    with open(input_fasta, "r") as infile, open(output_fasta, "w") as outfile:
        for record in SeqIO.parse(infile, "fasta"):
            try:
                protein_seq = record.seq.translate()
                protein_record = SeqRecord(protein_seq, id=record.id)
                SeqIO.write(protein_record, outfile, "fasta")
            except Exception as e:
                print(f"Error translating record {record.id}: {e}")
```
