```python
df = pd.read_csv('mcrA_results.txt', delimiter='\t')
df.columns = ['org_1', 'org_2', '%_identity', 'alignment_length', 'mismatches', 'gap_opens', 'start_1', 'end_1', 'start_2', 'end_2', 'e-value', 'bit_score']
df
#8724 rows 
```

```python
df = df[df['org_1'].str.strip() != df['org_2'].str.strip()]
#8432 rows
#strip of accession values into new columns
df[['accession_1', 'org_1']] = df['org_1'].str.split('|', 1, expand=True)
df[['accession_2', 'org_2']] = df['org_2'].str.split('|', 1, expand=True)
df = df.sort_values(by=['%_identity','org_1', 'org_2'])
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#df = df.reset_index(drop=True)
agg_df = df.groupby(['org_1', 'org_2'])['%_identity'].mean().reset_index()

heatmap_data = df.pivot(index='org_1', columns='org_2', values='%_identity')

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(data=heatmap_data, annot=True, cmap='YlGnBu', fmt='.1f', linewidths=.5)
plt.title('Percent Similarity Heatmap')
plt.xlabel('Species 2')
plt.ylabel('Species 1')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'species_1': ['A', 'A', 'B', 'C'],
    'species_2': ['X', 'X', 'Z', 'Z'],
    'percent_similarity': [0.8, 0.6, 0.9, 0.3]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Pivot the DataFrame
heatmap_data = df.pivot(index='species_1', columns='species_2', values='percent_similarity')

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
plt.title('Heatmap of Percent Similarity')
plt.show()
```
